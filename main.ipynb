```python
# Ensure df is defined before using it
from sklearn.datasets import load_digits
import pandas as pd

dataset = load_digits()
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)

x = df
y = dataset.target
```

```python
x = df
y = dataset.target
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size =0.2,random_state= 42)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
x #use PCA to reduce the dimensions of the data
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.95) #95% of the variance
x_pca = pca.fit_transform(x)
x_pca.shape
```

```python
pca.n_components_ #number of components to keep
```

```python
x_pca
```

```python
x_train_pca,x_test_pca,y_train,y_test = train_test_split(x_pca,y,test_size =0.2,random_state =42)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train_pca,y_train)
model.score(x_test_pca,y_test)
```

```python
pca =PCA(n_components=2)
x_pca =pca.fit_transform(x)
x_pca.shape
```

```python
x_pca
```

```python
x_train_pca,x_test_pca,y_train,y_test =train_test_split(x_pca,y,test_size=0.2,random_state=42)

model = LogisticRegression()
model.fit(x_train_pca,y_train)
model.score(x_test_pca,y_test)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt


pca = PCA(n_components=2)
x_pca = pca.fit_transform(x)
plt.scatter(x_pca[:,0],x_pca[:,1],c=y,edgecolor='none',alpha=0.5,s=40,cmap='viridis')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA of Digits Dataset')
plt.colorbar()
plt.show()
```
